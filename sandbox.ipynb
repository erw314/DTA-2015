```python
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle

%matplotlib inline
```

```python
#load file
f = open('driverData_0.p', 'r')
drivers = pickle.load(f)
```

```python
def tripPlot(n_driver, i_trip):

    colormap = [str((float(a)+1)/(len(i_trip)+1)) for a in range(len(i_trip))]
    
    for i in range(len(i_trip)):
        x = [a[0] for a in drivers[n_driver][i_trip[i]]]
        y = [a[1] for a in drivers[n_driver][i_trip[i]]]
    
        #plot in xy plane
        plt.plot(x, y)
        plt.scatter(x[0], y[0], color='g', s=20)
        plt.scatter(x[-1], y[-1], color='r', s=20)
        
    plt.show()
```

```python
plt.figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')
tripPlot(0, range(10))
```
