```python
import numpy as np
import matplotlib.pyplot as plt
import cPickle as pickle
from trip import Trip

%matplotlib inline
```

```python
#load file
f = open('driverData_small.p', 'r')
drivers = pickle.load(f)
```

```python
trip_obj = [Trip(0, drivers[0][i]) for i in range(200)]
```

```python
import transform as tf

sorted_trips = sorted([len(trip_obj[i].coordinates) for i in range(200)], reverse=True)
```

```python
cutoff_len = sorted_trips[int(0.2*len(sorted_trips))]
cutoff_len = 0
```

```python
tests = [Trip(0, tf.rotate_x_axis(trip_obj[i].coordinates)) for i in range(200) if len(trip_obj[i].coordinates) > cutoff_len]
```

```python
def distance(path1, path2):
    if len(path1) > len(path2):
        for i in range(len(path1)-len(path2)):
            path2.append(path2[-1])
    if len(path1) < len(path2):
        for i in range(len(path2)-len(path1)):
            path1.append(path1[-1])
    distance = sum([((path1[i][0]-path2[i][0])**2+(path1[i][1]-path2[i][1])**2)**0.5 for i in range(len(path1))])
    path1 = tf.reflect(path1)
    distanceR = sum([((path1[i][0]-path2[i][0])**2+(path1[i][1]-path2[i][1])**2)**0.5 for i in range(len(path1))])
    return min(distance,distanceR)
```

```python
true_examples = []

for i in range(len(tests)):
    for j in range(len(tests)):
        if (i != j) and abs(len(tests[i].coordinates)-len(tests[j].coordinates))/len(tests[i].coordinates)<0.2:
            if (distance(tests[i].coordinates,tests[j].coordinates)/len(tests[i].coordinates) < 20):
                if max([x[0] for x in tests[i].coordinates]) > 1000:
                    if i not in true_examples:
                        true_examples.append(i)
```
