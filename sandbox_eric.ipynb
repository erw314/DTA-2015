# Basic 3 feature analysis using outlier detection

```python
trip_obj = [Trip(1, i, drivers[0][i]) for i in range(200)]
```

The 3 basic features chosen are total distance, average speed, and average acceleration. We normalized each.

```python
def average_speed(obj):
    return np.mean([abs(obj.x_coordinates[i+1]-obj.x_coordinates[i]) for i in range(len(obj.x_coordinates)-1)])
```

```python
def average_acceleration(obj):
    speed = [abs(obj.x_coordinates[i+1]-obj.x_coordinates[i]) for i in range(len(obj.x_coordinates)-1)]
    return np.mean([abs(speed[i+1]-speed[i]) for i in range(len(speed)-1)])
```

```python
def normalize(stuff):
    r = max(stuff) - min(stuff)
    return [(x-min(stuff))/float(r) for x in stuff]
```

```python
normalize([-1, 2, 3])
```

```python
x = normalize([a.total_distance() for a in trip_obj])
y = normalize([average_speed(a) for a in trip_obj])
z = normalize([average_acceleration(a) for a in trip_obj])
plt.scatter(x, y)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
plt.show()
```

Here we use an elliptic envelope which estimates the parameters of an elliptical distribution, with number of outliers as parameter. We run the envelope for 0 to 50 outliers (out of 200 trips total) and calculate the percentage of outlier classifications for each trip. Those above a certain threshold are deemed to be 'bad' trips.

```python
from sklearn import covariance

points = np.array([[x[i]/1000, y[i], z[i]] for i in range(len(x))])

matrix = [[] for i in range(200)]

for i in range(0, 50):
    elliptic = covariance.EllipticEnvelope(contamination=float(i)/200)
    e = elliptic.fit(points)
    e = e.decision_function(points)
    for i in range(len(e)):
        matrix[i].append(e[i])
```

```python
threshold = 50
percent_neg = [sum([1 for i in range(len(n)) if n[i] < 0])/float(len(n))*100 for n in matrix]
bad = [i for i in range(len(percent_neg)) if percent_neg[i] > threshold]
print bad, len(bad)
```

The problem here is that the elliptical distribution ends up choosing outliers based heavily in some feature, in this case distance. Having to choose the number of outliers and the threshold is also problematic, as each heavily influence which are chosen to be outliers.

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
(ax.scatter([x[i] for i in range(len(x)) if i in bad],
            [y[i] for i in range(len(y)) if i in bad],
            [z[i] for i in range(len(z)) if i in bad]))
plt.show()
```

```python
plt.figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')
plt.xlim(-20000, 20000)
plt.ylim(-20000, 20000)
tripPlot(0, bad)

plt.figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k', )
plt.xlim(-20000, 20000)
plt.ylim(-20000, 20000)
tripPlot(0, [i for i in range(200) if i not in bad])
```

```python
import copy

def comp_dist(trip1, trip2):
    if len(trip1.x_coordinates) < len(trip2.x_coordinates):
        shorter = trip1
        longer = trip2
    else:
        shorter = trip2
        longer = trip1
    longer_x = longer.x_coordinates
    longer_y = longer.y_coordinates
    shorter_x = shorter.x_coordinates
    shorter_y = shorter.y_coordinates
    trip3 = Trip(0, tf.reflect(shorter.coordinates))
    shorter_rx = trip3.x_coordinates
    shorter_ry = trip3.y_coordinates
    
    iterations = len(longer_x)-len(shorter_y)
    all_dist = []
    all_rdist = []
    
    for i in range(iterations):
        # regular
        distance = [np.sqrt((longer_x[n+i]-shorter_x[n])**2+(longer_y[n+i]-shorter_y[n])**2) for n in range(len(shorter_x))]
        distance = sum(distance)/len(distance)
        all_dist.append(distance.copy())
        # reflected
        distance = [np.sqrt((longer_x[n+i]-shorter_rx[n])**2+(longer_y[n+i]-shorter_ry[n])**2) for n in range(len(shorter_x))]
        distance = sum(distance)/len(distance)
        all_rdist.append(distance.copy())
    
    return min(min(all_dist), min(all_rdist))
```

```python
total = 100
for i in range(total-1):
    for j in range(i+1, total):
        x = trip_obj[i].total_distance()
        y = trip_obj[j].total_distance()
        percent = (max(x, y)-min(x, y))/float(max(x, y))
        if percent < 0.2 and comp_dist(trip_obj[i], trip_obj[j]) < 100:
            print i, j
```
